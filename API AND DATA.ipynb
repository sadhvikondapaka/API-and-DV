```python
import requests
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# -----------------------------
# 1. Fetch data from OpenWeatherMap API
# -----------------------------
API_KEY = "your_api_key"   # Get a free API key from openweathermap.org
CITY = "Hyderabad"
URL = f"http://api.openweathermap.org/data/2.5/forecast?q={CITY}&appid={API_KEY}&units=metric"

response = requests.get(URL)
data = response.json()

# -----------------------------
# 2. Extract relevant weather data
# -----------------------------
if 'cod' in data and data['cod'] == '200':
    forecast_list = data['list']

    dates = []
    temps = []
    humidity = []
    weather_conditions = []

    for item in forecast_list:
        dates.append(item['dt_txt'])
        temps.append(item['main']['temp'])
        humidity.append(item['main']['humidity'])
        weather_conditions.append(item['weather'][0]['main'])

    # Convert to DataFrame
    df = pd.DataFrame({
        'Date': dates,
        'Temperature (°C)': temps,
        'Humidity (%)': humidity,
        'Condition': weather_conditions
    })

    # -----------------------------
    # 3. Data Visualization
    # -----------------------------
    plt.figure(figsize=(12,6))
    sns.lineplot(x="Date", y="Temperature (°C)", data=df, marker="o", label="Temperature")
    sns.lineplot(x="Date", y="Humidity (%)", data=df, marker="o", label="Humidity")
    plt.xticks(rotation=45)
    plt.title(f"Weather Forecast for {CITY}")
    plt.legend()
    plt.tight_layout()
    plt.show()

    # -----------------------------
    # 4. Condition Frequency Visualization
    # -----------------------------
    plt.figure(figsize=(6,4))
    sns.countplot(x="Condition", data=df)
    plt.title(f"Weather Conditions Frequency in {CITY}")
    plt.show()
else:
    print(f"Error fetching data: {data.get('message', 'Unknown error')}")
```
